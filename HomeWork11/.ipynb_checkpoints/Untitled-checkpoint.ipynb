```python
y = (sin(x)/2 + cos(x)/2)**2 * sin(x)
y
# Так выглядит функция в нормальном виде Jupyter Notebook:
```

## 1.Определить корни

```python
solveset(y, x)
```

```python
solve(y,x)
```

## 2.Найти интервалы, на которых функция возрастает

*2.1 Найдем производную функции:*

```python
derivate = diff(y)
derivate
```

*2.2 Определим период функции:*

```python
period = periodicity(y,x)
period
```

*2.3 Построим график призводной:*

```python
graph_dx_dt = plot(derivate)
р = graph_dx_dt
```

*2.4 Найдем точки экстремумов производной*

```python
extremum = solve(derivate, x)
for i in extremum:
    print(round(i.evalf(), 2), '\t', i)
```

*2.5 Отсортируем промежутки*

```python
if extremum[-1] != pi:
    extremum.append(pi)
if extremum[0] != pi:
    extremum.insert(0,-pi)
for i in extremum:
    print(round(i.evalf(), 2), i)
```

**2.6 Интервалы возрастания**

**3. И убывания функции**

```python
for i in range(len(extremum)-1):
    result = derivate.subs(x,(extremum[i+1]+ extremum[i])/2.0)
    if result > 0:
        print(f'Функция возрастает на интервале: ({round(extremum[i].evalf(), 2)}: {round(extremum[i+1].evalf(), 2)})')
    else:
        print(f'Функция убывает на интервале: ({round(extremum[i].evalf(), 2)}: {round(extremum[i+1].evalf(), 2)})')
```

## 4.Построить график (период)

## 5.Вычислить вершину

```python
y_max = maximum(y, x, Interval(-pi, pi)).evalf()
y_min = minimum(y, x, Interval(-pi, pi)).evalf()

print('y_max = ', round(y_max, 3))
print('y_min = ', round(y_min, 3))
```
